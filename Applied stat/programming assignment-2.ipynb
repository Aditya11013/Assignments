```python
import numpy as np
import matplotlib.pyplot as plt
import yfinance as yf
import pandas as pd
```

```python
apple= yf.download("AAPL",start='2020-01-01',end='2021-01-01')
```

```python
apple['Returns']=apple["Adj Close"].pct_change(1).dropna()
```

```python
apple['Returns']
```

```python
plt.figure(dpi=300,figsize=(9,3))
apple['Returns'].plot(kind='hist',bins=50)
plt.show()
```
