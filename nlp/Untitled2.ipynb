```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
# Let's scrape Trip Advisor
domain = "https://www.theoi.com/greek-mythology/greek-gods.html"
query = "olympian-gods"
URL = f"{domain}/{query}"
```

```python
response = requests.get(domain)
if response.status_code == 200:
    print("Successfully fetched webpage!")
else:
    print("Failed to fetch the webpage. Status code:", response.status_code)
```

```python
soup = bs(response.text, 'html.parser')
print(soup.body.prettify())
```

```python
links = soup.find_all("a")
links = [link['href'] for link in links if 'href' in link.attrs]
links
```

```python
from urllib.parse import urljoin
gods_link = [urljoin(domain, link) for link in links ]
gods_link = sorted(set(gods_link))
```

```python
len(gods_link)
```

```python
# Visit each link and extract details
data = []
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}
for link in gods_link:
    try:
        response = requests.get(link, headers=header)
        soup = bs(response.content, 'html.parser')

        # Extract title
        title = soup.find('title').get_text(strip=True) if soup.find('title') else "No Title"

        # Extract page text
        text = soup.get_text(separator=' ', strip=True)

        # Append to data
        data.append([title, link, text])
        print(f"Scraped: {title} | URL: {link}")
    except Exception as e:
        print(f"Error scraping {link}: {e}")
```

```python
data = []
header = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/110.0.0.0 Safari/537.36"
}

for link in gods_link:
    try:
        response = requests.get(link, headers=header)
        soup = bs(response.content, 'html.parser')

        # Extract title
        title = soup.find('title').get_text(strip=True) if soup.find('title') else "No Title"

        # Extract <p> elements
        paragraphs = soup.find_all('p')
        paragraph_text = " ".join(p.get_text(strip=True) for p in paragraphs)

        # Append to data
        data.append([title, link, paragraph_text])
        print(f"Scraped: {title} | URL: {link}")
    except Exception as e:
        print(f"Error scraping {link}: {e}")
```

```python
import os

# Create a folder in Colab to store the files
folder_name = "GreekGods"
if not os.path.exists(folder_name):
    os.makedirs(folder_name)

# Create files in the specified folder
for entry in data:
    title, url, info = entry  # Unpack the entry
    file_name = title.split(" -")[0].strip() + ".txt"  # Use text before '-' as the filename
    file_path = os.path.join(folder_name, file_name)  # Full path to save in the folder

    # Create formatted content
    content = f"Title: {title}\nURL: {url}\nInfo:\n{info}"

    # Write to a text file
    with open(file_path, "w", encoding="utf-8") as file:
        file.write(content)

    print(f"Created file: {file_path}")

# Zip the folder for easy downloading
import shutil
shutil.make_archive(folder_name, 'zip', folder_name)

# Download the zip file
from google.colab import files
files.download(f"{folder_name}.zip")
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
import re
```

```python
# Download NLTK resources
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Initialize tools
stop_words = set(stopwords.words('english'))
ps = PorterStemmer()
lemmatizer = WordNetLemmatizer()
```

```python
# Remove special characters and numbers
text = re.sub(r'[^a-zA-Z\s]', '', data[50][2])

# Remove extra whitespace
cleaned_text = re.sub(r'\s+', ' ', text).strip()
```

```python
# Sentence tokenization
sentences = sent_tokenize(cleaned_text)
# Word tokenization
words = word_tokenize(cleaned_text)
```

```python
# Remove stop words
filtered_words = [word for word in words if word.lower() not in stop_words]
```

```python
# Stemming
stemmed_words = [ps.stem(word) for word in filtered_words]

# Lemmatization
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
```

```python
# Output results
print("Cleaned Text:", cleaned_text)
print("Sentences:", sentences)
print("Filtered Words:", filtered_words)
print("Stemmed Words:", stemmed_words)
print("Lemmatized Words:", lemmatized_words)
```
